# ClinicalBERT PHI De-Identification NER

Model: `obi/deid_bert_i2b2`

This notebook uses a BERT token-classification model fine-tuned for clinical de-identification on i2b2-style PHI labels. It is useful for detecting protected health information, not medication or symptom entities.

Project usage note: use this notebook only for `AGE`. The remaining PHI categories should be recorded from `models/clinicalbert/clinicalbert_phi_ner_2.ipynb`.

## Used PHI Entities

For this project workflow, this notebook only keeps and displays this label:

| Entity | Meaning |
| --- | --- |
| `AGE` | Patient or person age |

The underlying model can predict additional PHI labels, but this notebook filters them out. Use `models/clinicalbert/clinicalbert_phi_ner_2.ipynb` for the remaining PHI categories.

The model uses BILOU-style labels internally: `B-`, `I-`, `L-`, `U-`, and `O`. The notebook aggregates these into readable entity spans.

```python
from __future__ import annotations

import pandas as pd
import torch
from transformers import AutoModelForTokenClassification, AutoTokenizer, pipeline

MODEL_NAME = "obi/deid_bert_i2b2"
device = 0 if torch.cuda.is_available() else -1

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForTokenClassification.from_pretrained(MODEL_NAME)

ner_pipeline = pipeline(
    task="token-classification",
    model=model,
    tokenizer=tokenizer,
    aggregation_strategy="simple",
    device=device,
)

print(f"Loaded {MODEL_NAME} on {'cuda' if device == 0 else 'cpu'}.")
print("Raw PHI labels:")
print(model.config.id2label)

ALLOWED_ENTITIES = {"AGE"}
print("Displayed PHI labels:", sorted(ALLOWED_ENTITIES))
```

Edit `input_text`, then run the detection cells.

```python
input_text = "On 03/12/2019 Dr. Smith at St. Mary's Hospital prescribed ibuprofen to Mr. Jones who 27 years old from india for severe knee pain. conatct aveenvas@gmail.com"
input_text
```

```python
def detect_phi_entities(text: str) -> pd.DataFrame:
    predictions = ner_pipeline(text)
    rows = []

    for prediction in predictions:
        entity = prediction.get("entity_group", prediction.get("entity"))
        if entity not in ALLOWED_ENTITIES:
            continue

        rows.append(
            {
                "entity": entity,
                "text": prediction["word"],
                "start": prediction["start"],
                "end": prediction["end"],
                "score": round(float(prediction["score"]), 4),
            }
        )

    return pd.DataFrame(rows, columns=["entity", "text", "start", "end", "score"])


entities = detect_phi_entities(input_text)
entities
```

```python
if entities.empty:
    print("No AGE entities detected.")
else:
    for row in entities.itertuples(index=False):
        print(f"{row.entity}: {row.text} ({row.score})")
```

Expected behavior on the default sentence: because this notebook only displays `AGE`, the output may be empty even if the underlying model detects other PHI labels such as `DATE`, `HOSP`, `PATIENT`, or `STAFF`.
